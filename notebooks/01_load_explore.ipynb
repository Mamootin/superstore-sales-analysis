```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append("..")
from src.data_prep import load_orders, clean_orders

df = clean_orders(load_orders("../data/raw/Superstore.xlsx"))
df.to_csv("../data/processed/orders_clean.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sales_by_region = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
sales_by_region.plot(kind="bar", title="Total Sales by Region")
plt.savefig("../reports/figures/sales_by_region.png", bbox_inches="tight")
plt.show()
```

```python
profit_by_category = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
profit_by_category.plot(kind="bar", title="Total Profit by Category")
plt.savefig("../reports/figures/profit_by_category.png", bbox_inches="tight")
plt.show()
```

```python
profit_by_subcategory = df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False)
profit_by_subcategory.plot(kind="bar", title="Total Profit by Sub-Category", figsize=(12, 5))
plt.savefig("../reports/figures/profit_by_subcategory.png", bbox_inches="tight")
plt.show()
```

```python
monthly_sales = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()
monthly_sales.index = monthly_sales.index.to_timestamp()
monthly_sales.plot(kind="line", title="Monthly Sales Trend", figsize=(12, 5))
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.savefig("../reports/figures/monthly_sales_trend.png", bbox_inches="tight")
plt.show()
```

Profit vs. discount scatter:

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["Discount"], df["Profit"], alpha=0.3)
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.title("Profit vs. Discount")
plt.axhline(0, color="red", linestyle="--", linewidth=1)
plt.savefig("../reports/figures/profit_vs_discount.png", bbox_inches="tight")
plt.show()
```

Top and bottom 10 products by profit:

```python
product_profit = df.groupby("Product Name")["Profit"].sum().sort_values(ascending=False)

top_10 = product_profit.head(10)
bottom_10 = product_profit.tail(10)

print("Top 10 most profitable products:")
print(top_10)
print("\nBottom 10 least profitable products:")
print(bottom_10)

top_10.to_csv("../reports/figures/top_10_products.csv")
bottom_10.to_csv("../reports/figures/bottom_10_products.csv")
```
